```python
new_columns = []
for i, col in enumerate(df.columns):
    if "Unnamed" in col:  # Unnamed 컬럼만 처리
        if i % 2 == 1:  # 홀수 번째
            new_columns.append("GDP 대비 온실가스배출량(톤CO2eq./만 달러)")
        else:  # 짝수 번째
            new_columns.append("1인당 온실가스배출량(톤 CO2eq./인)")
    else:
        new_columns.append(col)
df.columns = new_columns
```

```python
df
```

```python


# 데이터프레임 로드 (이미 제공된 데이터프레임을 사용)


# 1. 열 이름 정리
# 중복된 열 이름에서 연도를 추출하고, GDP 대비 배출량과 1인당 배출량을 구분
columns = df.columns
new_columns = []

for i, col in enumerate(columns):
    if i == 0:  # 첫 번째 열은 '국가' 그대로 유지
        new_columns.append('국가')
    else:
        year = col.split('.')[0]  # 연도 추출
        if i % 2 == 1:  # 홀수 번째는 'GDP 대비 배출량'
            new_columns.append(f"{year}_GDP 대비 배출량")
        else:  # 짝수 번째는 '1인당 배출량'
            new_columns.append(f"{year}_1인당 배출량")

# 새로운 열 이름 적용
df.columns = new_columns

# 2. 데이터 정리 (melt 사용)
# 국가를 고정하고 나머지 열을 '연도_지표'로 melt
df_melted = df.melt(id_vars=['국가'], var_name='연도_지표', value_name='값')

# 3. 연도와 지표 분리
# '연도_지표'를 연도와 지표로 나누기
df_melted['연도'] = df_melted['연도_지표'].str.extract(r'(\d{4})')  # 연도 추출
df_melted['지표'] = df_melted['연도_지표'].str.extract(r'(GDP 대비 배출량|1인당 배출량)')  # 지표 추출

```

2) 연도별 1인당 온실가스 배출량의 변화 추이는 어떻게 나타나는가?


```python
import seaborn as sns
import matplotlib.pyplot as plt
# '값' 열을 숫자로 변환 (변환할 수 없는 값은 NaN 처리)
df_melted['값'] = pd.to_numeric(df_melted['값'], errors='coerce')

# NaN 값 제거
df_melted = df_melted.dropna(subset=['값'])

# '1인당 배출량'만 필터링
per_capita_emission = df_melted[df_melted['지표'] == '1인당 배출량']

# 국가별 그래프 생성
unique_countries = per_capita_emission['국가'].unique()

for country in unique_countries:
    country_data = per_capita_emission[per_capita_emission['국가'] == country]

    # X축 순서를 0부터 재설정
    country_data = country_data.reset_index(drop=True)

    # 그래프 생성
    plt.figure(figsize=(8, 6))
    plt.plot(range(len(country_data)), country_data['값'], marker='o', linestyle='-', color='blue')  # X축: 순서
    plt.title(f'{country}의 1인당 온실가스 배출량 변화 추이', fontsize=16)
    plt.xlabel('순서', fontsize=12)
    plt.ylabel('1인당 온실가스 배출량 (톤 CO2eq./인)', fontsize=12)
    plt.grid(axis='y', linestyle='--', alpha=0.7)

    # y축 범위 설정
    y_min = country_data['값'].min()
    y_max = country_data['값'].max()
    plt.ylim(y_min - 1, y_max + 1)  # 여유를 두고 축 설정

    plt.show()

```

```python
# 값 열을 숫자로 변환
per_capita_emission['값'] = pd.to_numeric(per_capita_emission['값'], errors='coerce')

# NaN 값 제거 (숫자로 변환할 수 없는 값)
per_capita_emission = per_capita_emission.dropna(subset=['값'])

# 국가별 1인당 배출량 평균 계산
average_emission_by_country = per_capita_emission.groupby('국가')['값'].mean().sort_values(ascending=False)

# 막대 그래프 시각화
plt.figure(figsize=(12, 6))
average_emission_by_country.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('국가별 평균 1인당 온실가스 배출량', fontsize=16)
plt.xlabel('국가', fontsize=12)
plt.ylabel('평균 1인당 온실가스 배출량 (톤 CO2eq./인)', fontsize=12)
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

3) GDP 대비 배출량과 1인당 배출량 간 상관관계는 무엇인가?

```python
# 숫자로 변환하고, 변환할 수 없는 값은 NaN으로 처리
df_melted['값'] = pd.to_numeric(df_melted['값'], errors='coerce')

# NaN 값 제거
df_melted = df_melted.dropna(subset=['값'])
```

```python
gdp_vs_per_capita = df_melted.pivot_table(
    index=['국가'],  # 국가를 기준으로
    columns='지표',  # 'GDP 대비 배출량', '1인당 배출량'으로 구분
    values='값',  # 숫자로 변환된 값만 사용
    aggfunc='mean'  # 평균값 계산
).reset_index()

print(gdp_vs_per_capita.head())  # 결과 확인
# 상관관계 계산
correlation = gdp_vs_per_capita[['GDP 대비 배출량', '1인당 배출량']].corr().iloc[0, 1]

# 출력
print(f"상관계수: {correlation:.2f}")
```

```python
# 상관관계 행렬 계산
correlation_matrix = gdp_vs_per_capita[['GDP 대비 배출량', '1인당 배출량']].corr()

# 히트맵 시각화
plt.figure(figsize=(6, 5))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap="coolwarm", cbar=True, square=True)
plt.title('GDP 대비 배출량과 1인당 배출량 간 상관관계 히트맵', fontsize=12)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(data=gdp_vs_per_capita, x='GDP 대비 배출량', y='1인당 배출량', hue='국가')
plt.axhline(y=gdp_vs_per_capita['1인당 배출량'].mean(), color='red', linestyle='--', linewidth=1.5, label='1인당 배출량 평균')
plt.axvline(x=gdp_vs_per_capita['GDP 대비 배출량'].mean(), color='blue', linestyle='--', linewidth=1.5, label='GDP 대비 배출량 평균')
plt.title('GDP 대비 배출량과 1인당 배출량 간 상관관계 (평균선 포함)', fontsize=16)
plt.xlabel('GDP 대비 배출량 (톤 CO2eq./만 달러)', fontsize=12)
plt.ylabel('1인당 배출량 (톤 CO2eq./인)', fontsize=12)
plt.legend()
plt.grid(alpha=0.5)
plt.show()
```

# 결과 : 히트맵과 산점도 분석 결과, 두 변수 간의 상관계수는 약 -0.44로 음의 상관관계를 나타냄.
대부분의 국가가 GDP 대비 배출량과 1인당 배출량의 상관성을 보여주며, 선진국과 개발도상국의 그룹이 분리됨.
#알아낸 점
1.GDP 대비 온실가스 배출량이 높을수록 1인당 온실가스 배출량은 상대적으로 낮은 경향이 있음.
2. 개발도상국(예: 인도, 중국)은 경제 활동 대비 온실가스 배출량은 높으나, 1인당 배출량은 낮음.
3. 반대로 선진국(예: 미국, 호주)은 1인당 배출량이 높은 경향을 보이며 GDP 대비 배출량은 낮음.

#탐구할 점
1.GDP 대비 배출량이 낮고 1인당 배출량이 높은 국가에서의 배출 원인을 분석.
예: 미국과 호주의 높은 1인당 배출량의 주요 기여 요인(석탄, 석유 사용 등).
